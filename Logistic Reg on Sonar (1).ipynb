```python
clist = df.columns[0:60:1]
```

```python
def outliers(df):
    m = round(df.mean(),2)
    s = round(df.std(),2)
    low = round(m-(3*s),2)
    high = round(m+(3*s),2)
    df = df.map(lambda x : low if x < low else (high if x > high else x ))
    return df

```

```python
for i in clist:
    df[i] = outliers(df[i])
```
