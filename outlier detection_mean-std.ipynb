```python
import pandas as pd
import numpy as np
```

```python
m=np.random.randint(40,80,(50))
```

```python
m=np.insert(m,[3,5,19,25,30],[115,125,180,160,169])
```

```python
m
```

```python
avg=m.mean()
```

```python
avg=np.round(avg,2)
avg
```

```python
std=np.round(m.std(),2)
std
```

```python
lower=avg-(3*std)
lower=50
```

```python
higher=avg+(3*std)
higher=100
```

```python
lower
higher

```

# Trimming the Outliers..using mean std

```python
valid=m[(m>=lower) & (m<=higher)]
```

```python
valid.size
```

```python
m.size
```

```python
dm=pd.DataFrame(m)
```

# Capping the Outliers....Using Mean-std

```python
dm[0]=dm[0].apply(lambda x:lower if x<lower else higher if x>higher else x)
```

```python
dm
```
